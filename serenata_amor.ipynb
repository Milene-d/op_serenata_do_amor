```python
import pandas as pd
dados_1 = pd.read_csv("/content/despesa_ceaps_2022.csv", sep = ";", encoding="latin1", decimal = ",")
dados_2 = pd.read_csv("/content/despesa_ceaps_2021.csv", sep = ";", encoding="latin1", decimal = ",")
dados_3 = pd.read_csv("/content/despesa_ceaps_2020.csv", sep = ";", encoding="latin1", decimal = ",")
dados_4 = pd.read_csv("/content/despesa_ceaps_2019.csv", sep = ";", encoding="latin1", decimal = ",")
pd.set_option('mode.use_inf_as_na', True)
```

```python
dados = dados_1.append(dados_2)
```

```python
dados = dados.append(dados_3)
```

```python
dados = dados.append(dados_4)
```

Tratamento dos dados

```python
dados = dados.reset_index()
dados.head(5)
```

```python
#Retirar index
dados.columns = ['ANO', 'MES', 'SENADOR', 'TIPO_DESPESA', 'CNPJ_CPF',
                                           'FORNECEDOR', 'DOCUMENTO', 'DATA', 'DETALHAMENTO', 
                                           'VALOR_REEMBOLSADO', 'COD_DOCUMENTO', 'NA', 'NA']
```

```python
dados.info()
```

```python
dados.head(10)
```

```python
#Retirar primeira linha e duas últimas colunas
dados = dados[1:]
dados.head(5)
```

```python
dados = dados.drop(columns=['NA'])
dados.head(5)
```

```python
#Verificar tipo da coluna
dados.info()
```

```python
dados["ANO"] = pd.to_numeric(dados["ANO"], errors="coerce")
dados["MES"] = pd.to_numeric(dados["MES"], errors="coerce")
dados["VALOR_REEMBOLSADO"] = pd.to_numeric(dados["VALOR_REEMBOLSADO"], errors="coerce")
dados.info()
```

```python
#Remove duplicatas
dados.drop_duplicates(inplace = True)
dados.head()
```
